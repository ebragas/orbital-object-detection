```python
from PIL import Image, ImageDraw
from google.cloud import storage
from io import BytesIO
```

```python
!gcloud projects list
```

```python
client = storage.Client(project='reliable-realm-222318')
```

```python
coord = (1230, 320, 1310, 400)
```

```python
bucket = client.bucket('reliable-realm-222318-vcm')
blob = bucket.blob('pipeline/full/PSScene3Band_20180601_182249_103e_visual.tiff')
byte_string = blob.download_as_string()
image_bytes = BytesIO(byte_string)
image = Image.open(image_bytes)
```

```python
im = image.rotate(12, expand=True)
im = im.crop(im.getbbox())
```

```python
draw = ImageDraw.Draw(im)
draw.rectangle(coord, outline='red', width=3)
im.show()
```
